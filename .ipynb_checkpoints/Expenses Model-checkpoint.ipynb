```python
import datetime
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('expenses.csv')
df
```

```python
selected_columns =['expenseAmount', 'expenseDate']
df = df[selected_columns]
df.copy()
df
```

```python
df['expenseDate'] = pd.to_datetime(df['expenseDate'], dayfirst=True)
df.index = df['expenseDate']
df
```

```python
df['Year'] = df['expenseDate'].dt.year
df['Month'] = df['expenseDate'].dt.month
df_monthly = df.groupby(['Year', 'Month'])['expenseAmount'].sum().reset_index()
# df_monthly = df_monthly.set_index('Month')
df_monthly
```

```python
expenses = df_monthly['expenseAmount']
expenses.plot()
```

```python
def df_to_X_y(df, window_size=5):
    df_as_np = df.to_numpy()
    X = []
    y = []
    for i in range(len(df_as_np) - window_size):
        row = [[a] for a in df_as_np[i:i+5]]
        X.append(row)
        label = df_as_np[i+5]
        y.append(label)
    return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X, y = df_to_X_y(expenses, WINDOW_SIZE)
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
X_train, y_train = X[:30], y[:30]
X_val, y_val = X[30:34], y[30:34]
X_test, y_test = X[34:], y[34:]
X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model = Sequential()
model.add(InputLayer((5,1)))
model.add(LSTM(64))
model.add(Dense(8, 'relu'))
model.add(Dense(1, 'linear'))

model.summary()
```

```python
cp = ModelCheckpoint('model1/', save_best_only=True)
model.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=1.0), metrics = [RootMeanSquaredError()])
```

```python
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=1000, callbacks=[cp])
```

```python
from tensorflow.keras.models import load_model
model = load_model('model1/')
```

```python
train_predictions = model.predict(X_train).flatten()
train_results = pd.DataFrame(data={'Train Predictions': train_predictions, 'Actuals':y_train})
train_results
```

```python
val_predictions = model.predict(X_val).flatten()
val_results = pd.DataFrame(data={'Val Predictions': val_predictions, 'Actuals':y_val})
val_results
```

```python
test_predictions = model.predict(X_test).flatten()
test_results = pd.DataFrame(data={'Train Predictions': test_predictions, 'Actuals':y_test})
test_results
```
